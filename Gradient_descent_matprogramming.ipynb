```python
#Gradient Descent

import os
from numpy import *
from matplotlib.pyplot import *
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import scipy.optimize as optimize
```

```python
def f(x,y):
    return 100*(y - x**2)**2+(1-x)**2
```

```python
# (X,Y) grid on [-20,20]
x = arange(-1, 1, 0.1)
y = arange(-1, 1, 0.1)
X, Y = meshgrid(x, y)

Z = f(X,Y)
```

```python
# Plot (X,Y) in 3D
fig = figure(1)
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.RdBu,linewidth=0, antialiased=False)



ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
fig.colorbar(surf, shrink = 0.7, aspect=5)
show()
```

```python
def f0(point):
    dx = 2*(-1+point[0]+200*point[0]**3-200*point[0]*point[1])
    dy = 200*(-point[0]**2+point[1])
    return dx, dy

def main():
    # set the learning rate
    lrate = 0.002
    # initialize a point
    a = np.array([-1, 1])
    # set number of epochs
    epoch = 100000
    result = []
    for i in range(epoch):
        # objective function
        f = (a[0]**2) + (100*((a[1] - a[0]**2)**2))
        result.append([a,f])
        # compute its derrivative on point a
        # derrivative
        fi = np.array(f0(a))
        a = a - np.dot(lrate,fi)

    # convert into array
    result = np.array(result)
    print(result[-10:-1])

    #printing out result at several points
    print(f'the minimum is: {result[-1, 1]} at point: {result[-1,0]} at point {result[-1,1]}')

main()
```

```python
# 3. Minimize the functions with Python minimize function and compare results
#define the function
def f(params):
   x,y = params
   return 100*(y-x**2)**2+(1-x)**2

# starting values
ig = [-1, 1]

#optimizing the function through optimize method
result = optimize.minimize(f, ig)

#printing out result
print(result.x)
```
