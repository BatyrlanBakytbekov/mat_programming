```python
# importing of necessary libraries for minimization and plotting
# scipy, numpy, matpotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import scipy.optimize as optimize
```

```python
# rule to find minimum of a function
# Steps:
# Find the derivative of the function and equate it to zero.
# Find the roots of the differentiated equation.
# Do double differentiation of original function and substitute the values of roots in the 2nd differentiated expression.
# If the value of double derivative after substituting the root is positive, Minimum occurs. Then substitute the value in original equation to get Minimum value of the function.
```

```python
#task 1 Find minimum of one variable function

# function: f(x)= x^6/6 - x^3 +2x

#define the function
def f(x):
  return x**6 *1/6 -x**3 +2*x

#plotting
x = np.arange(-2,2,0.1)
plt.plot(x, f(x))
plt.show()

#minimizing through minimize method from scipy
local_min = minimize(f, x0=1)
global_min = minimize(f, x0=0)

#printing out results
print(local_min)
print("\nglobal minimum: ", global_min.x)
print("local minimum: ",local_min.x, "\nfunction: ",local_min.fun)
```

```python
# function: f(x)= x + 1/exp(x-1) - 1

#define the function
def f(x):
 return x + (1/np.exp(x - 1) - 1)

#plotting
x = np.arange(-1,1,0.1)
plt.plot(x,f(x))
plt.show()

#minimizing through minimize method from scipy
result = minimize(f, x0=2)

#printing out results
print(result)
print("\nlocal minimum: ",result.x, "\nfunction: ",result.fun)
```

```python
# function: f(x)= x - sin(x)

#define the function
def f(x):
 return x - np.sin (x)

#plotting
x = np.arange(-10,10,0.1)
plt.plot(x,f(x))
plt.show()

#minimizing through minimize method from scipy
result = minimize(f, x0=2)

#printing out results
print(result)
print("\nlocal minimum: ",result.x, "\nfunction: ",result.fun)
```

```python
# task 2. Analyse two variable functions
# function f(x,y) = x^2+y^2

#define the function
def f(params):
   x,y = params
   return x**2+y**2

# starting values
ig = [1, 1]

#optimizing the function through optimize method
result = optimize.minimize(f, ig)

#printing out result
print(result.x)
```

```python
# Defense work
# function f(x,y) = 2x^2 + 2y^2

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import scipy.optimize as optimize
from numpy import exp,arange
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt

#define the function
def f(x,y):
  return 2*x**2+2*y**2

#plotting
x = arange(-1.5, 1.5, 0.1)
y = arange(-1.5, 1.5, 0.1)

# grid of point
X,Y = meshgrid(x, y)
F = f(X, Y)

#drawing the function
im = imshow(F,cmap=cm.RdBu)

#contour lines
cset = contour(F,arange(-1.5,1.5,0.1),linewidths=2,cmap=cm.Set2)
clabel(cset,inline=True,fmt='%1.1f',fontsize=10)

#title
title("f(x,y) = 2x^2 + 2y^2")
show()

# 3d plotting of a function
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, F, rstride=1, cstride=1, cmap=cm.RdBu,linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

title("3d plot")
plt.show()



def fun(params):
  x,y = params
  return 2*x**2+2*y**2

# minimization
x0 = [3, 4]
result = optimize.minimize(fun, x0, options={'disp': True, "maxiter": 25})

print(result)
print("\nglobal minimum is: ",result.x)
```
