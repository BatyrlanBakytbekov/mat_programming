```python
import os
from numpy import *
from matplotlib.pyplot import *
import numpy as np
from scipy.linalg import cholesky, solve_triangular, cho_solve, cho_factor
from scipy.linalg import solve
from scipy.optimize import minimize
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import scipy.optimize as optimize
from numpy import exp,arange
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
#task1
def f(x):
    return (x[0]-1)**2 + 5*(x[1]-1)**2

# starting values
ig = [-1, 1]

cons = ({'type': 'eq',
         'fun' : lambda x: np.array([(x[0] + 1)**2+x[1]])})

#optimizing the function through minimize method
result = minimize(f, ig, constraints=cons)

#printing out result
print("minimum:")
print(result.x)
print(result.fun)
```

```python
#define the function
def f(x,y):
  return (x-1)**2+5*(y-1)**2

#plotting
x = arange(-1.5, 1.5, 0.1)
y = arange(-1.5, 1.5, 0.1)

# grid of point
X,Y = meshgrid(x, y)
F = f(X, Y)

#drawing the function
im = imshow(F,cmap=cm.RdBu)

#contour lines
cset = contour(F,arange(-1.5,1.5,0.1),linewidths=2,cmap=cm.Set2)
clabel(cset,inline=True,fmt='%1.1f',fontsize=10)

#title
title("f(x,y) = (x+1)^2+y")
show()

# 3d plotting of a function
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, F, rstride=1, cstride=1, cmap=cm.RdBu,linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

title("3d plot")
plt.show()
```

```python
#task 2
def f(x):
    return 100*(x[1]-x[0]**2)**2 + (1-x[0])**2

# starting values
ig = [-1, 1]

cons = ({'type': 'eq',
         'fun' : lambda x: np.array([(x[0] + 1)**2+x[1], x[0]**2+x[1]**2-1.5])})

#optimizing the function through minimize method
result = minimize(f, ig, constraints=cons)

#printing out result
print("minimum:")
print(result.x)
print(result.fun)
```

```python
#define the function
def f(x,y):
  return 100*(y-x**2)**2 + (1-x)**2

#plotting
x = arange(-1.5, 1.5, 0.1)
y = arange(-1.5, 1.5, 0.1)

# grid of point
X,Y = meshgrid(x, y)
F = f(X, Y)

#drawing the function
im = imshow(F,cmap=cm.RdBu)

#contour lines
cset = contour(F,arange(-1.5,1.5,0.1),linewidths=2,cmap=cm.Set2)
clabel(cset,inline=True,fmt='%1.1f',fontsize=10)

#title
title("f(x1,x2) = 100*(x2-x1**2)**2 + (1-x1)**2")
show()

# 3d plotting of a function
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, F, rstride=1, cstride=1, cmap=cm.RdBu,linewidth=0, antialiased=False)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

title("3d plot")
plt.show()
```
